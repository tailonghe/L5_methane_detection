# Caclualte leak rates and cumulative emissions

```python
det_num_df = pd.read_csv('figure_reproduce/plume_num.csv')
cloud_corr_df = pd.read_csv('figure_reproduce/cov_adj_num.csv')
ems_df = pd.read_csv('figure_reproduce/ems.csv')
ems_err_df = pd.read_csv('figure_reproduce/ems_err.csv')
cf_num_df = pd.read_csv('figure_reproduce/cf_num.csv')



det_num_sub = det_num_df[(det_num_df['Longitude']-53.833<=0.05)   &  (det_num_df['Latitude']-39.391<=0.05 )]
cloud_corr_sub = cloud_corr_df[(cloud_corr_df['Longitude']-53.833<=0.05)   &  (cloud_corr_df['Latitude']-39.391<=0.05 )]
ems_sub = ems_df[(ems_df['Longitude']-53.833<=0.05)   &  (ems_df['Latitude']-39.391<=0.05 )]
ems_err_sub = ems_err_df[(ems_err_df['Longitude']-53.833<=0.05)   &  (ems_err_df['Latitude']-39.391<=0.05 )]
cf_num_sub = cf_num_df[(cf_num_df['Longitude']-53.833<=0.05)   &  (cf_num_df['Latitude']-39.391<=0.05 )]


det_num_sub
```

```python
likely = det_num_sub.sum().values[2:]/cf_num_sub.mean().values[2:]

fig = pplt.figure(refwidth=5, refheight=2.5, sharex=False, sharey=False)
gs = pplt.GridSpec(1, 2, wspace=0, hspace=0)

ax1 = fig.add_subplot(gs[0, :], ycolor='#0077BB', ylabel='Number of detections')
ax1.number = 7
ax2 = ax1.twinx(yloc=('axes', 1), color='#CC3311', alpha=0.1, label='Percentage of clear-sky images', ylim=(0, 100))
ax3 = ax1.twinx(yloc=('axes', 1.12), color='#009988', alpha=0.1, label='Cumulative emissions (Tg)', ylim=(0, 1))
likely[likely > 1] = 1
ax1.axvspan(2002, 2008, facecolor='gray', hatch='//', edgecolor='white', alpha=0.3)
ax2.plot(np.arange(1986, 2012, 1), likely*100, color='#CC3311', alpha=0.3, lw=3)
ax2.fill_between(np.arange(1986, 2012, 1), 0, likely*100, color='#CC3311', alpha=0.05, lw=0)
ax1.bar(np.arange(1986, 2012, 1), det_num_sub.sum().values[2:], color='#0077BB', alpha=0.4)


aems = ems_sub.sum().values[2:]
aems_err = ems_err_sub.sum().values[2:]
aems = np.nancumsum(aems)/1e3
aems_err = np.nancumsum(aems_err)/1e3
upperb = aems+aems_err
lowerb = aems-aems_err
ax3.plot(np.arange(1986, 2012, 1)[:-9], aems[:-9], color='#009988', lw=3)
ax3.fill_between(np.arange(1986, 2012, 1)[:-9], lowerb[:-9], upperb[:-9], color='#009988', alpha=0.1)

ax3.plot(np.arange(1986, 2012, 1)[-4:], aems[-4:], color='#009988', lw=3)
ax3.fill_between(np.arange(1986, 2012, 1)[-4:], lowerb[-4:], upperb[-4:], color='#009988', alpha=0.1)

fig.format(abc='A', abcloc='ul', abcsize=20)
fig.format(xlim=(1986, 2011))
plt.grid(False)
# fig.savefig('fig3b_Aug9.pdf')
```
